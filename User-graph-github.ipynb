The idea is to check whether the person , you are following, is following you back or not on the github.

### Problem Statement :
A github user, Watson, follows 3 gihub users Allo, Cortana and Alexa. Watson wants to find out whether Allo Alexa and Cortana follow him back or not.

### What python module will we need ?
In this we will be extracting data from Github rest endpoints. To pull data from endpoints/urls we are going to use <b>requests</b> module.

```python
import requests
```

<u><b><i>Check people deepali99, is following</i></b></u>

```python
def getUserFollowing(user):
    '''Returns a list of users (json objects) which the username is following'''
    base_url = 'https://api.github.com/users'
    type_resource = 'following'
    new_url = base_url + '/' + user + '/' + type_resource
    #print(new_url)
    response = requests.get(new_url)
    username_is_following = response.json()
    print (username_is_following)
    return username_is_following
```

```python
def followback(user):
    #I am treating user as myself
    my_circular_followings = []
    #get all follwings of the user
    my_followings = getUserFollowing(user)
    #my_followings contains the list of users (json) follwed by me
    
    #let's find followings of each users in the my_followings list
    for my_following in my_followings:
        #my_following is  a json data contaning informtaion of one of the user I am follwing 
        
        #let's get the name of the user I am follwing from my_following json data
        my_following_login_name = my_following['login']
        
        #let's get the following of the user (my_following_login_name) I am following
        my_following_followings = getUserFollowing(my_following_login_name)
        
        #let's iterate in the list of my followings' followings (my_following_followings) 
        #to check whether my following follows me or not
        for my_following_following in my_following_followings:
            my_following_following_login_name = my_following_following['login']
            if(user == my_following_following_login_name ):
                print(user)
                print(my_following_following_login_name)
                print('circular Following is success')
                my_circular_followings.append(my_following_login_name)
            else:
                print(user)
                print(my_following_following_login_name)
                print('circular Following is failure')
    return my_circular_followings
```
